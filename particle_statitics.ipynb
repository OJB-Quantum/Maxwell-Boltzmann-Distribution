```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Constants
k_B = 8.617333262145e-5  # Boltzmann constant in eV/K
T = 0  # Temperature in Kelvin. Try out 0K, 100K, and 500K.
E_F = 1.0  # Fermi level (in eV)

# Energy mesh
E_min = 0
E_max = 2
NE = 1000
energy = np.linspace(E_min, E_max, NE)  # Energy range in eV

# Distributions
def fermi_dirac(energy, T, E_F):
    return 1 / (np.exp((energy - E_F) / (k_B * T)) + 1)

def maxwell_boltzmann(energy, T):
    return np.exp(-(energy - E_F) / (k_B * T))

def bose_einstein(energy, T):
    with np.errstate(divide='ignore', invalid='ignore'):
        be_distribution = 1 / (np.exp((energy - E_F) / (k_B * T)) - 1)
        be_distribution[energy == 0] = np.inf  # Handle the case where energy equals zero
    return be_distribution

# Calculate distributions
fd_distribution = fermi_dirac(energy, T, E_F)
# mb_distribution = maxwell_boltzmann(energy, T)
# be_distribution = bose_einstein(energy, T)
```

```python
# Plot distributions
plt.figure(figsize=(10, 6))
plt.plot(energy, fd_distribution, label='Fermi-Dirac', color='blue')
# plt.plot(energy, mb_distribution, label='Maxwell-Boltzmann', color='green')
# plt.plot(energy, be_distribution, label='Bose-Einstein', color='red')

# Labels and Title
plt.xlabel('Energy (eV)')
plt.ylabel('Distribution Function')
plt.title(f'Distribution Functions at T = {T} K, $E_F$ = {E_F} eV')
plt.legend()

# Set y-axis limits from 0 to 3
plt.ylim(0, 3)
# Show plot
plt.grid(True)
plt.show()
```

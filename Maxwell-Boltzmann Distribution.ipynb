```python
# Maxwell-Boltzmann Distribution

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.pyplot import figure

m1 = 132*1.66054e-27

T1 = 300
kb = 1.38064852e-23
alpha1 = np.sqrt(2*kb*T1/m1)

v = np.arange(0,1000,1)
Pv1 = (4/(np.sqrt(np.pi)*alpha1**3))*(v**2)*np.exp(-(v/alpha1)**2)
Flux1 = v*Pv1

m2 = 40*1.66054e-27

T2 = 300
kb = 1.38064852e-23
alpha2 = np.sqrt(2*kb*T2/m2)


Pv2 = (4/(np.sqrt(np.pi)*alpha2**3))*(v**2)*np.exp(-(v/alpha2)**2)
Flux2 = v*Pv2

plt.figure(figsize=(8,5), dpi=500)

plt.plot(v,Pv1,label = '$Au (Recommended)$', color='blue',lw=3)
plt.plot(v,Pv2,label = '$Au (So-Called-Precise)$', linestyle = '--', color='red',lw=3)
plt.title('Maxwell-Boltzmann Distribution',fontdict={'fontname':'DejaVu Sans','fontsize':20})
plt.xlabel('Velocity (m/s)')
plt.ylabel('Flux')


plt.grid(which='both', color = 'blue', linestyle = '--', linewidth = 0.5)
plt.show()
```

```python
# Maxwell-Boltzmann Distribution

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.pyplot import figure
import math

e = 2.718281828459045
pi = 3.14159265359

m1 = 0.197

T1 = 300
r = 8.314
u1 = (3/2)
## u1 is just a variable I made to represent the fractional exponent in the equation. 

v = np.arange(0,1000,1)
# v = np.linspace(0,1000,100)
# -abs(some #) is used to force a number to become negative.

F1 = 4*pi*(((m1)/((2*pi)*(r)*(T2)))**u1)*(v**2)*(e**(((-abs(m1))*((v)**2))/((2)*(r)*(T2))))*100


m2 = 0.029
T2 = 300

F2 = 4*pi*(((m2)/((2*pi)*(r)*(T2)))**u1)*(v**2)*(e**(((-abs(m2))*((v)**2))/((2)*(r)*(T2))))*100


plt.figure(figsize=(8,5), dpi=500)

plt.plot(v,F1, color='blue',lw=3)
plt.plot(v,F2, linestyle = '--', color='red',lw=3)
plt.title('Maxwell-Boltzmann Distribution',fontdict={'fontname':'DejaVu Sans','fontsize':20})
plt.xlabel('Velocity (m/s)')
plt.ylabel('Percent of Molecules at Each Velocity [%/(m/s)]')


plt.grid(which='both', color = 'blue', linestyle = '--', linewidth = 0.5)
plt.show()
```

```python
# Chamber Evacuation Time

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.pyplot import figure
import math

e = 2.718281828459045
pi = 3.14159265359

Sp1 = 0.707 ## this is the pump speed in Liters per second
Pp1 = 2*(10**(-3))  ## this is minimum pressure
Ip = 760 ## this is the initial pressure
V = 10 ## this is the volume in cubic meters
t = np.arange(0,60,1)
# t = np.linspace(0,60,10000)
# -abs(some #) is used to force a number to become negative.

#P1 = e**(-abs((Sp1)*(t))/(V))
P1 = (((Ip-Pp1)*(e**((-abs(Sp1)*(t))/V))+(Pp1))/(t))

Sp2 = 3.767 
Pp2 = 5.03*(10**(-4))

#P2 = e**(-abs((Sp2)*(t))/(V))
P2 = (((Ip-Pp2)*(e**((-abs(Sp2)*(t))/V))+(Pp2))/(t))


plt.figure(figsize=(8,5), dpi=500)

plt.plot(t,P1, color='blue',lw=3)
plt.plot(t,P2, color='red',lw=3)
plt.title('Chamber Evacuation Time',fontdict={'fontname':'DejaVu Sans','fontsize':20})
plt.xlabel('Time (min)')
plt.ylabel('Pressure (Torr)')

#plt.yscale('symlog')
plt.grid(which='both', color = 'blue', linestyle = '--', linewidth = 0.5)
plt.show()
```
